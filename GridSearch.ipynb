### KNN on Titanic Dataset 

```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head(3)
```

```python
def change(x):
    if(x == 'male'):
        return 0
    else:
        return 1
```

```python
set1 = data[['PassengerId','Sex', 'Parch','Fare']].copy()
```

```python
set1.head()
```

```python
set1['Sex'] = set1['Sex'].apply(change)
```

```python
set1
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_data = set1.copy()
```

```python
output_data = data[['Survived']].copy()
```

```python
output_data.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, x_test, Y, y_test = train_test_split(training_data, output_data, test_size=0.3, random_state = 42)
```

```python
model = KNeighborsClassifier(n_neighbors=6)
```

```python
model.fit(X, Y)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test, prediction)
score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion = confusion_matrix(prediction, y_test)
confusion
```
